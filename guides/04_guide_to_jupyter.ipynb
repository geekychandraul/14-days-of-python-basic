## ðŸ§  Explaining the *Kernel*

Behind every Jupyter notebook is a **Python process** that runs your code â€” this process is called the **Kernel**.

Each notebook has its **own dedicated Kernel**, which keeps track of all the variables, functions, and imports you've run.

You can control the Kernel from the top menu by going to **`Kernel â†’ Restart Kernel`**. This will stop the current Python process and start a fresh one â€” effectively wiping all memory of variables and previous code executions.

If you restart the Kernel and immediately try to run the next cell, youâ€™ll likely see an error. Thatâ€™s because the variable `favorite_fruit` is no longer defined (NameError) â€” it's been lost with the restart.

To fix this, simply **run all the cells from the top of the notebook**, in order. Once you've done that, the cell will execute without issue.

> ðŸ” Restarting the Kernel is useful when things get messy or if you're not sure whatâ€™s been run and what hasnâ€™t.

## What is a NameError?
A `NameError` in Python occurs when you try to use a name (a variable, function, class, module, etc.) that Python does not recognize. In simpler terms:

> A NameError means Python does not know what youâ€™re talking about.

There are a few common reasons for a NameError:

1. **You spelled a name incorrectly** (typos are the most frequent cause).

2. **You referenced a name before it was defined** (i.e., you tried to use a variable, function, or class before its definition).

3. **You havenâ€™t imported a module that youâ€™re trying to use** (e.g., math.sqrt() without import math).

4. **You used a name in the wrong scope** (e.g., using a local variable outside of the function in which it was declared).

## How to Debug a NameError
1. **Check for typos**: Make sure the name you typed matches exactly what is defined.
 
    - Python is case-sensitive. `myFunction` and `myfunction` are different names.

2. **Check the order of statements**: Ensure you define a name (variable, function, class, or import) **before** using it.

3. **Check whether youâ€™ve run the definition** (in a Jupyter notebook, you must run the cell that defines a name before using it in any other cell).

4. *Ensure your imports are correct*: Did you do `import something` or `from something import something_else`?

## ðŸ”¹ Minor Things We Encounter in the Course

This next item isn't a built-in feature of Jupyter itself, but it's a **handy Python package** that's especially useful when working in JupyterLab â€” and you'll see it used throughout this course.

The package is called **`tqdm`**, and it adds a **progress bar** to any iterable. Itâ€™s great for tracking the progress of loops, especially when working with large datasets or long-running processes.

Just wrap your iterable with `tqdm(...)` and it will automatically display a progress bar in the output cell.

> ðŸ’¡ Bonus: It works beautifully in Jupyter Notebooks and is super helpful for debugging or just staying patient while code runs.

You can install it with:

```python
!pip install tqdm
```

````python
# Here's some code with no progress bar
# It will take 10 seconds while you wonder what's happpening..

import time

vars = ["var"] * 1000

for var in vars:
    time.sleep(0.01)
````

````python
# And now, with a nice little progress bar:

import time
from tqdm import tqdm

vars = ["var"] * 1000

for var in tqdm(vars):
    time.sleep(0.01)
````

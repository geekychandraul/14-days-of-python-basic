## üó£Ô∏è Topic 1: The `print()` Function
We need a way to see what our code is doing. In Python, we use `print()`.
Think of `print()` as a megaphone. Whatever you put inside the parentheses `( )`, Python shouts it out to the screen.

````python
print("Hello, World!")
print("I am writing Python code!")
print("Please don't crash.")
````

### üß† The "Quotes" Rule (Crucial!)
Notice how we put text inside `"double quotes"`?
* **With Quotes:** Python treats it as **Text** (a String). It blindly prints whatever is inside.
* **Without Quotes:** Python treats it as **Code** (commands, numbers, or variables).

**Let's try a science experiment:**
Run the cell below. Why does one look like math and the other look like text?

````python
print("2 + 2") 
print(2 + 2)
````

**The Result:**
* `"2 + 2"` is just text. Python is dumb; it just prints the symbols.
* `2 + 2` (no quotes) is code. Python activates its brain, calculates the math, and prints `4`.

## üí• Topic 2: Dealing with Errors (The Red Text of Doom)
You will make mistakes. Lots of them.
When you see **Red Text**, do not panic. The computer is not angry at you; it's just confused.

**Common Mistake #1: The Missing Quote**
Run this and see what happens.

````python
# This will crash. Can you guess why?
print("I love Python)
````

**How to fix it:**
Python is saying *SyntaxError: EOL while scanning string literal*.
In human language: *"Bro, you started a quote but never finished it. I don't know where the text ends!"*

**Task:** Go back up and add the missing `"` to fix the code.

## ü§´ Topic 3: Comments (The Secret Notes)
Sometimes you want to write notes to yourself, like *"Fix this later"* or *"I have no idea why this works."*
You don't want Python to try and execute those notes.

We use the hash symbol `#` for comments. Python ignores everything after a `#`.

````python
# This is a comment. Python can't see me.
print("Python can see this!") 

# print("I am invisible code")
````

## ðŸ”„ Topic 1: The Syntax
It looks exactly like an `if` statement, but it keeps repeating as long as the condition is `True`.



[Image of while loop flow chart]


**Warning:** If the condition *never* becomes False, the loop runs forever. This is called an **Infinite Loop**. (Your computer fan might start screaming).

```python
# A simple counter
count = 1

while count <= 5:
    print(f"Count is: {count}")
    count += 1  # âš ï¸ CRITICAL: We must change 'count', or the loop never ends!

print("Finished!")
```

### The "Infinite Loop" Panic

**The Scenario**: You might run while x < 5: print(x) but forget x += 1. The computer will flood with 0s.

**The Fix**: You need to use the Kill Switch immediately.

**In Jupyter**: Click the "Stop" (Square) button at the top.

**In Terminal**: Press Ctrl + C.

#### While vs For (When to use what?)

**Analogy**:

Use for when you know How Many times (Loop 10 times, Loop over every item in a list).

Use while when you know the Goal but not the time (Loop until the user says 'stop', Loop until Game Over).

---
## ðŸ›‘ Topic 2: The Emergency Brake (`break`)
Sometimes you want to exit a loop immediately, even if the condition is still True.
We use the keyword `break`.

This is super useful for **"Do this forever until I say stop"** situations.

```python
while True: # This creates an Infinite Loop on purpose!
    command = input("Type 'exit' to stop: ")
    
    if command == "exit":
        print("Stopping the loop. Goodbye!")
        break  # The Eject Button
    
    print(f"You typed: {command}")
```

---
## â­ï¸ Topic 3: The Skip Button (`continue`)
What if you want to skip *just this one round* of the loop, but keep going?
We use `continue`. It jumps back to the top of the loop immediately.

```python
# Print numbers from 1 to 5, but skip 3.
n = 0
while n < 5:
    n += 1
    
    if n == 3:
        print("Skipping 3!")
        continue  # Go back to start of loop. Don't run code below.
        
    print(f"Number is {n}")
```

#### The continue Confusion

Beginners rarely use continue correctly. They usually put it after the print statement, which makes it useless.

Tip: Show them that continue means "Teleport back to the top". It skips everything under it.

## ðŸ‘‹ Welcome Back!
Imagine you have a phone book.
* **The List Way:** You have a list of names `["Alice", "Bob"]` and a separate list of numbers `[1234, 5678]`. If you sort one list, the other one gets messed up. Nightmare.
* **The Dictionary Way:** You look up "Alice" and find "1234".

A **Dictionary** stores data in **Key-Value Pairs**.
* **Key:** The label (e.g., "Name").
* **Value:** The data (e.g., "Tony Stark").



---

## ðŸ”‘ Topic 1: Creating a Dictionary
We use **Curly Brackets** `{}` (just like Sets), but we use a colon `:` to separate Key and Value.
`{ Key : Value }`

````python
# A simple superhero profile
hero = {
    "name": "Bruce Wayne",
    "alias": "Batman",
    "age": 40,
    "is_rich": True
}

print(hero)
````

---
## ðŸ” Topic 2: Accessing Data (The Lookup)
Lists use numbers `[0]`. Dictionaries use **Keys** `["name"]`.

````python
print(hero["name"])  # Bruce Wayne
print(hero["alias"]) # Batman

# âŒ This will CRASH if the key doesn't exist
# print(hero["power_level"]) # KeyError!

# âœ… Safer Way: .get()
# If the key is missing, it returns 'None' instead of crashing.
print(hero.get("power_level"))
````

#### JSON Connection

**Know this**: The syntax for Dictionaries { "key": "value" } is almost identical to JSON, which is how the entire Internet sends data. If you master Dictionaries, you master the Web.

---
## â™»ï¸ Topic 3: Modifying Data
Dictionaries are **Mutable** (Changeable).
* You can **Add** new keys.
* You can **Update** existing keys.

````python
student = {"name": "Harry", "house": "Gryffindor"}

# 1. ADD a new key
student["pet"] = "Hedwig"
print("After Add:", student)

# 2. UPDATE an existing key
student["house"] = "Slytherin" # Traitor!
print("After Update:", student)

# 3. DELETE a key
student.pop("pet")
print("After Delete:", student)
````

#### The KeyError (The Dictionary Crash)

**Scenario**: Student types student["Name"] (uppercase 'N') when the key is "name".

**The Crash**: KeyError: 'Name'.

**The Fix**: Keys are case-sensitive strings. Dictonary has  `.get\()` method is a "Safety Net."

```python
student.get("Name")
```

## ðŸ” Topic 4: Looping through Dictionaries
You can loop through Keys, Values, or Both.

````python
prices = {"Apple": 2, "Banana": 1, "Cherry": 5}

# Loop through Keys (Default)
print("--- Fruits ---")
for fruit in prices:
    print(fruit)

# Loop through Values
print("--- Prices ---")
for cost in prices.values():
    print(cost)

# Loop through Both (Most useful!)
print("--- Menu ---")
for fruit, cost in prices.items():
    print(f"The {fruit} costs ${cost}")
````

---
## ðŸŽ Topic 5: Nested Dictionaries (Inception)
You can put a Dictionary *inside* a Dictionary. This is how complex data (like Instagram posts) looks.

````python
users = {
    "user1": {
        "name": "Alice",
        "age": 25
    },
    "user2": {
        "name": "Bob",
        "age": 30
    }
}

# How to access Bob's age?
# Go to 'user2', then go to 'age'
print(users["user2"]["age"])
````

#### Why Dictionaries > Lists?

**Analogy**:

- **List**: A pile of books. To find "Harry Potter," you have to check every single book until you find it (Slow).

- **Dictionary**: A Library Index. You go to "H," find "Harry Potter," and grab it immediately (Fast).

**Time Complexity**: Dictionaries are O(1) (Constant Time lookup), while Lists are O(n) (Linear Time).

## ðŸ”¢ Topic 6: Dictionary Functions (Cheat Codes)

### Methods in Dictionary 

| Method | Description | Example |
|--------|-------------|---------|
| `clear()` | Removes all items from the dictionary | `d={'a':1}; d.clear()` â†’ `{}` |
| `copy()` | Returns a shallow copy of the dictionary | `d={'a':1}; d2=d.copy()` â†’ `{'a':1}` |
| `fromkeys(seq[, value])` | Creates a new dict with keys from `seq` and given value | `dict.fromkeys(['a','b'], 0)` â†’ `{'a':0,'b':0}` |
| `get(key[, default])` | Returns value for key, or `default` if not found | `d={'a':1}; d.get('b',0)` â†’ `0` |
| `items()` | Returns a view object of (key, value) pairs | `d={'a':1}; d.items()` â†’ `dict_items([('a',1)])` |
| `keys()` | Returns a view object of dictionary keys | `d={'a':1}; d.keys()` â†’ `dict_keys(['a'])` |
| `pop(key[, default])` | Removes key and returns its value (or `default`) | `d={'a':1}; d.pop('a')` â†’ `1` |
| `popitem()` | Removes and returns the last inserted (key, value) pair | `d={'a':1,'b':2}; d.popitem()` â†’ `('b',2)` |
| `setdefault(key[, default])` | Returns value if key exists, else inserts with default | `d={}; d.setdefault('a',5)` â†’ `5`, dict â†’ `{'a':5}` |
| `update([other])` | Updates dict with key/value pairs from `other` | `d={'a':1}; d.update({'b':2})` â†’ `{'a':1,'b':2}` |
| `values()` | Returns a view object of dictionary values | `d={'a':1}; d.values()` â†’ `dict_values([1])` |


#### The .items() Unpacking
`.items()` will be most used function with dictionary when running loops. 

`.items()` returns a tuple (key, value), and we are unpacking it into two variables instantly. It's Python magic.

## ðŸ” Topic 2: Indexing (Grabbing Items)
Just like Strings, Lists use an **Index** starting at **0**.

| Iron Man | Thor | Cap | Hulk |
| :---: | :---: | :---: | :---: |
| 0 | 1 | 2 | 3 |

```python
heroes = ["Iron Man", "Thor", "Captain America", "Hulk"]

print(heroes[0])  # Iron Man
print(heroes[2])  # Captain America

# Negative Indexing (From the end)
print(heroes[-1]) # Hulk
```

#### The IndexError (The Classic Crash)

**Scenario**: A student creates a list with 3 items ([0, 1, 2]) and tries to access list[3].

**The Error**: IndexError: list index out of range.

**Explanation**: "You tried to grab the 4th item from a box that only has 3 items. Python panic."

## â™»ï¸ Topic 3: Mutability (The Edit Button)
Yesterday, we learned that Strings are **Immutable** (Change-proof).
Lists are **Mutable**. You can change them freely!

We can add, remove, or swap items.

```python
movies = ["Batman", "Superman", "Spiderman"]

# 1. SWAP: Changing an item
movies[1] = "Wonder Woman" 
print("After Swap:", movies)

# 2. APPEND: Adding to the END (Most common!)
movies.append("Flash")
print("After Append:", movies)

# 3. INSERT: Adding to a specific spot
movies.insert(0, "Aquaman") # Puts Aquaman at the very front
print("After Insert:", movies)

# 4. Append using cancatination
movies = movies + ["Ant-Man"]
print(f"After Concatination: {movies}")
```

#### `.append()` vs `list = list + item`

Beginners often try `list = list + "New Item"`. This crashes because you can't add List + String.

You can do `list = list + ["New Item"]` (List + List), but `.append()` is the professional way. Stick to `.append()` for now.

## ðŸ—‘ï¸ Topic 4: Removing Items
There are two main ways to delete things.

1.  `.pop(index)`: Removes by **Position** (Index). If blank, removes the last one.
2.  `.remove(value)`: Removes by **Name** (Value).

```python
fruits = ["Apple", "Banana", "Cherry", "Date"]

# Method 1: Pop (The Index Killer)
# Let's remove Banana (Index 1)
print(f"Fruit basket has: {fruits}")
removed_item = fruits.pop(1)
print(f"I ate the {removed_item}")
print("Leftovers:", fruits)

# Method 2: Remove (The Search & Destroy)
# Let's remove "Date"
fruits.remove("Date")
print("Final List:", fruits)
```

#### `pop()` vs `remove()`

**Confusion**: "Which one do I use?"

**Guideline**:

- Use `pop()` when you know WHERE it is (The last one, the first one).

- Use `remove()` when you know WHAT it is ("Remove 'Mushrooms'").

## ðŸ”¢ Topic 5: List Functions (Cheat Codes)
Python gives us tools to analyze number lists instantly.

| Function | What it does |
| :--- | :--- |
| `len(list)` | Counts total items |
| `max(list)` | Finds highest number |
| `min(list)` | Finds lowest number |
| `sum(list)` | Adds everything up |

```python
grades = [55, 90, 80, 100, 30]

print(f"Total Students: {len(grades)}")
print(f"Highest Score: {max(grades)}")
print(f"Lowest Score: {min(grades)}")
print(f"Class Average: {sum(grades) / len(grades)}")
```

### List Methods
> NOTE: You can use dir(list) to get this list. 

| Method | Description | Example |
|--------|-------------|---------|
| `append(x)` | Adds an item to the end of the list | `lst = [1,2]; lst.append(3)` â†’ `[1,2,3]` |
| `extend(iterable)` | Appends all elements from another iterable | `lst=[1]; lst.extend([2,3])` â†’ `[1,2,3]` |
| `insert(i, x)` | Inserts `x` at index `i` | `lst=[1,3]; lst.insert(1,2)` â†’ `[1,2,3]` |
| `remove(x)` | Removes first occurrence of `x` | `lst=[1,2,2]; lst.remove(2)` â†’ `[1,2]` |
| `pop([i])` | Removes and returns item at index `i` (last if omitted) | `lst=[1,2,3]; lst.pop()` â†’ returns `3`, list â†’ `[1,2]` |
| `clear()` | Removes all items from list | `lst=[1,2]; lst.clear()` â†’ `[]` |
| `index(x[, start[, end]])` | Returns first index of `x` | `lst=[1,2,3,2]; lst.index(2)` â†’ `1` |
| `count(x)` | Returns number of times `x` appears | `lst=[1,2,2,3]; lst.count(2)` â†’ `2` |
| `sort(key=None, reverse=False)` | Sorts the list in place | `lst=[3,1,2]; lst.sort()` â†’ `[1,2,3]` |
| `reverse()` | Reverses the list in place | `lst=[1,2,3]; lst.reverse()` â†’ `[3,2,1]` |
| `copy()` | Returns a shallow copy of the list | `lst=[1,2]; new=lst.copy()` â†’ `[1,2]` |

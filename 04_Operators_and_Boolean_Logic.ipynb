## ‚öñÔ∏è Topic 1: Comparison Operators (The Interrogation)
We use these symbols to ask Python a question. Python replies with True or False.

**‚ö†Ô∏è CRITICAL RULE:**
* `a = 5`: **Assignment** (Make `a` equal to 5).
* `a == 5`: **Question** (Is `a` equal to 5?).

| Symbol | Meaning | Example | Result |
| :--- | :--- | :--- | :--- |
| `==` | Equal to | `5 == 5` | True |
| `!=` | Not Equal to | `5 != 3` | True |
| `>` | Greater than | `10 > 2` | True |
| `<` | Less than | `10 < 2` | False |
| `>=` | Greater or Equal | `18 >= 18` | True |
| `<=` | Less or Equal | `5 <= 1` | False |

```python
# Let's ask Python some questions
print(10 > 5)   # True
print(10 == 10) # True
print(10 == 11) # False

# The "Not Equal" operator (!=)
# "Is 5 NOT equal to 3?" -> Yes, that is true.
print(5 != 3)
```

#### The "Single Equals" vs "Double Equals" Battle

**The Struggle**: This is the hardest concept for beginners. They will type if x = 5 and get an error.

**The Analogy**:

`=` (Assignment): A label maker. "I command you to be 5."

`==` (Comparison): A balance scale. "Are you the same weight as 5?"

### ‚ö†Ô∏è String Sensitivity
Python is very strict with text.

```python
print("Apple" == "apple") 
# Result: False (Capital 'A' is not small 'a')

print("10" == 10)
# Result: False (Text is not a Number)
```

#### Case Sensitivity

**Scenario**: A student types "Admin" and it fails.

**Explanation**: "Computers are dumb. To a computer, 'A' is a specific binary code (01000001) and 'a' is different (01100001). They are not related. They are strangers."

## ‚ö° Topic 2: Assignment Shortcuts (The Lazy Coder)
Before we get to complex logic, here is a quick trick.
Programmers hate typing `score = score + 1`. We use shortcuts.

| Long Way | Shortcut | Meaning |
| :--- | :--- | :--- |
| `x = x + 1` | `x += 1` | Add 1 to x |
| `x = x - 5` | `x -= 5` | Subtract 5 from x |
| `x = x * 2` | `x *= 2` | Double x |

```python
score = 0
score += 10  # Score is now 10
score += 5   # Score is now 15
score -= 2   # Score is now 13
print(f"Final Score: {score}")
```

#### The "Pythonic" Range

**Advanced Tip**: Python is one of the rare languages that allows 13 <= age <= 19.

The standard way that works in all language like Java/C++ is (age >= 13 and age <= 19). But Python also has a shortcut which is a "cool trick."

## üß† Topic 3: Logical Operators (And, Or, Not)
Real life decisions are complex.
*"I will go to the party if I have money **AND** if I am invited."*

We use 3 words to combine logic:



### 1. The `and` Operator
**Both** sides must be True.
* True + True = **True**
* True + False = **False**

```python
has_ticket = True
has_id = False

# Can I enter the concert?
print(has_ticket and has_id)
# Result: False (You need BOTH)
```

### 2. The `or` Operator
**Only one** side needs to be True.
* True + True = **True**
* True + False = **True**

*"I will eat if I am hungry OR if there is pizza."*

```python
is_hungry = False
is_pizza = True

# Will I eat?
print(is_hungry or is_pizza)
# Result: True (I'm not hungry, but it's pizza... so yes)
```

### 3. The `not` Operator
It simply **flips** the result.
* not True $\rightarrow$ False
* not False $\rightarrow$ True

```python
is_raining = True
is_sunny = not is_raining
print(is_sunny) # False
```

#### Boolean Variables are weird

Beginners struggle to understand that is_teen is a variable that holds True or False.

They often try to write logic like: print(f"Is teen: {if age > 13}").

**Clarification**: The comparison age > 13 transforms into the value True before it gets stored.

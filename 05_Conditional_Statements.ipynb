## üõ§Ô∏è Topic 1: The `if` Statement
The `if` statement checks a condition. If the condition is **True**, it runs the code block. If **False**, it skips it.

### ‚ö†Ô∏è The Golden Rule: Indentation
Python does not use `{}` curly brackets like other languages. It uses **Indentation (Spaces)**.
* Everything indented (pushed to the right) belongs to the `if` block.
* Usually, this is **4 spaces** or **1 Tab**.

````python
age = 20

if age >= 18:
    print("You are an adult.")
    print("You can vote!")

print("This line runs no matter what because it is NOT indented.")
````

#### The "IndentationError" Nightmare

**Expectation**: Few of the students will get an error today because they mixed spaces and tabs, or forgot to indent.

**The Fix**: See the red line in VS Code. Explaination: "In Python, whitespace is not empty space. It is grammar. It tells the code who belongs to whom.". Ideal situation use only tab which is equivalent to 4 spaces in code editors.

## ‚ÜîÔ∏è Topic 2: The `else` Statement
What if the condition is False? We need a backup plan.
The `else` keyword means: **"If the condition above was False, do THIS instead."**

````python
password = input("Enter password: ")

if password == "secret123":
    print("Access Granted.")
    print("Welcome, Agent.")
else:
    print("Access Denied!")
    print("Calling the police...")
````

---
## ü™ú Topic 3: The `elif` Statement (Multiple Choices)
Life isn't always Yes/No. Sometimes there are many options.
* "If Red, stop."
* "Else If (`elif`) Yellow, slow down."
* "Else, Go."

You can have as many `elif` blocks as you want!

````python
score = int(input("Enter your score (0-100): "))

if score >= 90:
    print("Grade: A (Amazing!)")
elif score >= 80:
    print("Grade: B (Good job)")
elif score >= 70:
    print("Grade: C (You passed)")
else:
    print("Grade: F (See me after class)")
````

### üïµÔ∏è Logic Check
Why didn't we check `if score < 90 and score >= 80` for Grade B?
* Because Python checks from **Top to Bottom**.
* If `score` was 95, the first `if` catches it, runs the code, and **exits** the whole block. It never even looks at the `elif`.

#### The if vs elif Trap

**Concept**: Why can't I just use three if statements?

**Demonstration**:

```Python
# BAD CODE
score = 95
if score > 80: print("A")
if score > 70: print("B")  # This will ALSO print because 95 > 70!
```
**Explanation**: elif connects them. It says "Only check me if the previous guy failed."

## üéÅ Topic 4: Nested Conditionals (The Inception)
You can put an `if` inside another `if`. This is called **Nesting**.

````python
has_ticket = True
has_id = False

if has_ticket:
    # We only check for ID if they have a ticket
    if has_id:
        print("Welcome to the show!")
    else:
        print("You need an ID to enter.")
else:
    print("You need to buy a ticket first.")
````

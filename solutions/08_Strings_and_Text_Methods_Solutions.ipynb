```python
# Level 1: Initials Extractor

f_name = input("First Name: ")
l_name = input("Last Name: ")
# Grab index 0 from both
print(f"{f_name[0]}.{l_name[0]}.")
```

```python
# Level 2: The Slicer

sentence = "Python is amazing"
# "is" starts at 7, ends at 9
word1 = sentence[7:9]
# "amazing" starts at 10 to end
word2 = sentence[10:]

print(word1)
print(word2)
```

```python
# Level 3: The Censor Machine

text = input("Say something: ")
# Case sensitivity matters! Ideally convert to lower first, but simple replace works too.
clean_text = text.replace("math", "****")
print(clean_text)
```

```python
# Level 4: The Palindrome Checker (Interview Classic)

word = input("Enter a word: ").lower() # Convert to lower to handle "Madam"
reversed_word = word[::-1]

if word == reversed_word:
    print("Yes, it is a palindrome!")
else:
    print("No, it is not.")
```

#### The Palindrome Logic (Level 4)

This is a very popular interview question.

**Explaination [::-1]**

- start (empty) = Beginning

- stop (empty) = End

- step (-1) = Walk backwards.

```python
# Level 5: The Email Parser

raw_email = "  USER@Example.com  " # Or use input()

# Chaining methods! .strip() happens first, then .lower()
clean_email = raw_email.strip().lower()

print(f"Cleaned Email: {clean_email}")

# Bonus: Check validity
if clean_email.endswith(".com"):
    print("Valid .com domain")
```

#### Chaining Methods (Level 5)

See `text.strip().lower()`.

It works left-to-right. strip() returns a string, and then .lower() runs on that result. This is "Pythonic" coding.

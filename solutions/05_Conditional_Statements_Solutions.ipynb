```python
# Level 1: Positive or Negative

num = int(input("Enter number: "))
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# Level 2: The Voting Booth

age = int(input("Enter age: "))
if age >= 18:
    print("You can vote!")
else:
    diff = 18 - age
    print(f"Too young. Come back in {diff} years.")
```

```python
# Level 3: Odd/Even Checker

num = int(input("Enter number: "))
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Level 4: The Rollercoaster (Nested)

height = int(input("Height in cm: "))

if height > 120:
    print("You can ride!")
    age = int(input("What is your age? "))
    if age < 12:
        print("Please pay $5.")
    else:
        print("Please pay $12.")
else:
    print("Sorry, you have to be taller to ride.")
```

```python
# Level 5: Text Adventure (Simple Version)

print("Welcome to Treasure Island.")
choice1 = input("You are at a crossroad. Where do you go? Type 'left' or 'right': ").lower()

if choice1 == "left":
    choice2 = input("You come to a lake. Type 'wait' for boat or 'swim': ").lower()
    if choice2 == "wait":
        print("You arrived at the island safely. YOU WIN!")
    else:
        print("You got eaten by a shark. GAME OVER.")
elif choice1 == "right":
    print("You fell into a hole. GAME OVER.")
else:
    print("You stood there until you starved. GAME OVER.")
```

#### The .lower() trick (Level 5)

In the solution for Level 5, I used .lower().

Students will test their game, type "Left" (Capital L), and lose immediately because "Left" != "left".

Learning: "You can force the user's input to lowercase to make your game robust. Try adding .lower() to your input! `"Left".lower()`"

```python
# Level 1: Bucket List

bucket_list = []
bucket_list.append("Skydiving")
bucket_list.append("Visit Japan")
bucket_list.append("Learn Python")
print(bucket_list)
```

```python
# Level 2: The Replacement

toppings = ["Pepperoni", "Mushrooms", "Onions"]
toppings[1] = "Cheese"
print(toppings)
```

```python
# Level 3: VIP Guest List

guests = ["Tom", "Jerry", "Mickey"]
guests.remove("Jerry") # Bye Jerry
guests.append("Donald") # Hi Donald
print(guests)
```

```python
# Level 4: Stats Machine

prices = [10.99, 5.50, 20.00, 3.25, 50.00]
print(f"Cheapest: ${min(prices)}")
print(f"Expensive: ${max(prices)}")
print(f"Total: ${sum(prices)}")
```

```python
# Level 5: List from User (Crucial Pattern!)

user_numbers = []

for i in range(3):
    num = int(input("Enter a number: "))
    user_numbers.append(num)

print(f"Your List: {user_numbers}")
print(f"Sum: {sum(user_numbers)}")
```

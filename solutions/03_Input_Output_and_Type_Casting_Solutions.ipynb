```python
# Level 1: The Chatterbox

color = input("What is your favorite color? ")
print(f"I love {color} too!")
```

```python
# Level 2: The Square Maker

num = int(input("Enter a number: "))
print(f"The square is {num ** 2}")
```

```python
# Level 3: Pizza Splitter (Floor Division)

slices = int(input("Total slices: "))
people = int(input("Total people: "))
per_person = slices // people
print(f"Each person gets {per_person} slices.")
```

```python
# Level 4: The Leftovers (Modulo)

# Assuming 'slices' and 'people' exist from Level 3
leftover = slices % people
print(f"The host gets the remaining {leftover} slices.")
```

```python
# Level 5: Hours & Minutes Converter

total_mins = int(input("Enter total minutes: "))
hours = total_mins // 60
minutes = total_mins % 60

print(f"{total_mins} minutes is {hours} Hours and {minutes} Minutes.")
```

```python
# Level 1: Birthday Singer

def happy_birthday(name):
    print("Happy Birthday to you")
    print(f"Happy Birthday dear {name}")
    print("Happy Birthday to you!")

happy_birthday("Alice")
```

```python
# Level 2: Math Machine
def multiply(x, y):
    result = x * y
    print(f"{x} multiplied by {y} is {result}")

multiply(5, 5)
```

```python
# Level 3: ID Card Generator

def make_id(name, age, school="Hogwarts"):
    print(f"Name: {name} | Age: {age} | School: {school}")

make_id("Harry", 11)
make_id("Sheldon", 25, "Caltech")
```

```python
# Level 4: Loop Inside Function

def print_times(text, loops):
    for i in range(loops):
        print(text)

print_times("I love Python", 3)
```

```python
# Level 5: Calculator App

def calculate(a, b, operator):
    if operator == "+":
        print(a + b)
    elif operator == "-":
        print(a - b)
    elif operator == "*":
        print(a * b)
    else:
        print("Unknown Operator")

calculate(10, 2, "+")
calculate(10, 2, "-")
```

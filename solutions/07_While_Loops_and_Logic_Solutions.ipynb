```python
# Level 1: Blast Off

count = 10
while count > 0:
    print(count)
    count -= 1
print("Blast Off!")
```

```python
# Level 2: The Annoying Kid

response = ""
while response != "yes":
    response = input("Can I have candy? ")
print("Thank you!")
```

```python
# Level 3: Password Gatekeeper

secret = "python"
while True:
    guess = input("Enter password: ")
    if guess == secret:
        print("Access Granted")
        break
    else:
        print("Wrong, try again.")
```

```python
# Level 4: The Odd Skipper
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue # Skip the even numbers
    print(n)
```

```python
# Level 5: Guess the Number

secret_number = 7

while True:
    guess = int(input("Guess a number (1-10): "))
    
    if guess == secret_number:
        print("You won! ðŸŽ‰")
        break
    elif guess < secret_number:
        print("Too low!")
    else:
        print("Too high!")
```

```python
# Level 1: Basic Truths

print(50 > 20)           # True
print("Python" == "Java") # False
print(10 != 10)          # False
```

```python
# Level 2: Smart Light Switch

is_daytime = True
lights_on = not is_daytime
print(f"Lights on? {lights_on}")
```

```python
# Level 3: The Teenager Check

age = int(input("Enter age: "))
# Option 1: The standard way
is_teen = (age >= 13) and (age <= 19)

# Option 2: The Pythonic way (Very cool to show students!)
# is_teen = 13 <= age <= 19

print(f"Is Teenager: {is_teen}")
```

```python
# Level 4: The Game Store (Shortcuts)

money = 100
money -= 40  # 60 left
money += 20  # 80 total
print(f"Final Wallet: ${money}")
```

```python
# Level 5: Secure Login System

correct_user = "admin"
correct_pass = "1234"

# Note: We don't convert password to int, because passwords can have letters!
u_name = input("Username: ")
p_word = input("Password: ")

# The Logic
success = (u_name == correct_user) and (p_word == correct_pass)
print(f"Access Granted: {success}")
```

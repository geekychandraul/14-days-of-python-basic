```python
# --- CONFIGURATION ---
balance = 1000  # Starting money
correct_pin = 1234
transaction_history = [] # Bonus: Keep track of actions!
```

```python
def check_balance():
    print(f"üí∞ Current Balance: ${balance}")

def deposit(amount):
    global balance # We need permission to change the global variable
    balance += amount
    print(f"‚úÖ Deposited ${amount}")
    transaction_history.append(f"Deposited ${amount}")
    check_balance()

def withdraw(amount):
    global balance
    if amount > balance:
        print("‚ùå Error: Insufficient Funds!")
    else:
        balance -= amount
        print(f"‚úÖ Withdrew ${amount}")
        transaction_history.append(f"Withdrew ${amount}")
        check_balance()
```

```python
def login():
    attempts = 3
    while attempts > 0:
        try:
            pin = int(input("Enter 4-digit PIN: "))
            
            if pin == correct_pin:
                print("üîì Access Granted!")
                return True # Success!
            else:
                attempts -= 1
                print(f"‚ùå Wrong PIN. Attempts left: {attempts}")
                
        except ValueError:
            print("‚ùå Please enter numbers only.")
            
    print("üö´ ACCOUNT LOCKED.")
    return False # Failed
```

```python
def start_atm():
    # 1. Try to Login First
    if login() == False:
        return # Stop the program if login fails

    # 2. Start the Menu Loop
    while True:
        print("\n--- üèß MAIN MENU ---")
        print("1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. View History")
        print("5. Exit")
        
        try:
            choice = int(input("Select an option (1-5): "))
            
            if choice == 1:
                check_balance()
            elif choice == 2:
                amt = float(input("Enter deposit amount: "))
                deposit(amt)
            elif choice == 3:
                amt = float(input("Enter withdraw amount: "))
                withdraw(amt)
            elif choice == 4:
                print("\nüìú Transaction History:")
                for log in transaction_history:
                    print(log)
            elif choice == 5:
                print("üëã Thank you for banking with Python!")
                break # Kill the loop
            else:
                print("‚ùå Invalid Option. Try 1-5.")
                
        except ValueError:
            print("‚ùå Error: Please enter a valid number.")

# --- RUN THE APP ---
start_atm()
```

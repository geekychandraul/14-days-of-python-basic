```python
# Level 1: The Safe Printer

items = ["A", "B", "C"]
try:
    print(items[5])
except IndexError: # Good habit to be specific, but 'except:' works too
    print("Index out of range!")
```

```python
# Level 2: The Age Verifier

try:
    age = int(input("Enter Age: "))
    print(f"Your age is {age}")
except ValueError:
    print("Please use digits like 20.")
```

```python
# Level 3: The Smart Divider

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"

print(safe_divide(10, 2))
print(safe_divide(5, 0))
```

```python
# Level 4: The Dictionary Safeguard

prices = {"apple": 2, "banana": 1}
item = input("What do you want? ").lower()

try:
    print(f"Price: ${prices[item]}")
except KeyError:
    print("Item not found in menu.")
```

```python
# Level 5: The Indestructible Loop (Very Common Pattern)

while True:
    try:
        num = int(input("Please enter a number: "))
        print(f"Thank you, you entered {num}")
        break # Exit loop only if line above succeeds
    except ValueError:
        print("That's not a number! Try again.")
```

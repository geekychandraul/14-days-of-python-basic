```python
# Level 1: Name Repeater

name = input("Enter name: ")
count = int(input("How many times? "))

for i in range(count):
    print(name)
```

```python
# Level 2: Pattern Printer
 
for num in range(1, 11):
    square = num ** 2
    print(f"{num} * {num} = {square}")
```

```python
# Level 3: Evens Only

total = 0 # This is our empty bucket

for number in range(1, 101): # 1 to 100
    total += number # Add the number to the bucket

print(f"The sum of 1 to 100 is: {total}")
```

```python
# Level 4: Factorial Calculator

num = int(input("Enter number: "))
result = 1 # Must start at 1, because 0 * anything = 0!

for i in range(1, num + 1):
    result *= i

print(f"The factorial of {num} is {result}")
```

```python
# Level 5: FizzBuzz (The Boss Level)

for number in range(1, 51):
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)
```

#### FizzBuzz Logic Order (Level 5)

**The Trap**: You might have put the `if number % 3 == 0` check first.

If the number is 15:

It prints "Fizz" and the elif/else chain stops.

It never checks for "FizzBuzz".

**The Fix**: "Always check the most specific/strict condition first (Divisible by BOTH)!"

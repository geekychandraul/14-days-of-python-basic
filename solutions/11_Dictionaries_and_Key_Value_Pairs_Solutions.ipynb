```python
# Level 1: Contact Card

contact = {
    "name": "Tony Stark",
    "phone": "555-0100",
    "email": "tony@stark.com"
}
print(f"Call {contact['name']} at {contact['phone']}")
```

```python
# Level 2: Grade Update

grades = {"Math": 80, "English": 70}
grades["Math"] = 95       # Update
grades["Science"] = 85    # Add
print(grades)
```

```python
# Level 3: The Translator

spanish = {"hello": "hola", "thank you": "gracias", "yes": "si"}
word = input("Enter word (hello/yes): ").lower()

# .get() prevents crashing if they type "goodbye"
print(f"Translation: {spanish.get(word)}")
```

```python
# Level 4: Receipt Printer

menu = {"Burger": 5, "Fries": 3, "Soda": 2}
total = 0

print("--- Receipt ---")
for item, price in menu.items():
    print(f"{item} ...... ${price}")
    total += price

print(f"----------------\nTotal: ${total}")
```

```python
# Level 5: Student Database (Nested)

classroom = {
    "student_1": {"name": "John", "gpa": 3.5},
    "student_2": {"name": "Jane", "gpa": 3.8}
}

# Print Jane's GPA
print(classroom["student_2"]["gpa"])

# Update John's GPA
classroom["student_1"]["gpa"] = 4.0
print(classroom["student_1"])
```

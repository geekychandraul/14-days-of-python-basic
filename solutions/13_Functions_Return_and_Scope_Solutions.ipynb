```python
# Level 1: The Square Machine

def square(number):
    return number * number

my_sq = square(5)
print(f"Square is: {my_sq}")
```

```python
# Level 2: Currency Converter

def usd_to_eur(dollars):
    return dollars * 0.85

wallet = usd_to_eur(100)
print(f"You have â‚¬{wallet}")
```

```python
# Level 3: Area Calculator

def calc_area(length, width):
    return length * width

room = calc_area(10, 10)   # 100
garden = calc_area(5, 5)   # 25

total_area = room + garden # We can do math because they are Numbers!
print(f"Total Area: {total_area}")
```

```python
# Level 4: Title Formatter

def format_name(f_name, l_name):
    # .title() makes the first letter Capital, rest lowercase
    full_name = f"{f_name.title()} {l_name.title()}"
    return full_name

clean_name = format_name("tONY", "stARK")
print(clean_name)
```

```python
# Level 5: The Life Counter (Global)

lives = 3

def hit():
    global lives # Permission to edit
    lives -= 1
    print("Ouch! Lost a life.")

hit()
hit()
print(f"Lives remaining: {lives}")
```

```python
# Level 2: The Year Calculator
# Goal: Basic integer math.

current_year = 2025
birth_year = 2003

age = current_year - birth_year
print(age)
```

```python
# Level 3: The Price Tag (Floats)
# Goal: Float math and F-strings.

price = 999.99
tax_rate = 0.10  # 10%

tax_amount = price * tax_rate
total_price = price + tax_amount

# We use :.2f inside the curly braces to round to 2 decimal places!
print(f"The total price is ${total_price:.2f}")
```

```python
# Level 4: The Variable Swap (Logic Puzzle)
# Goal: Understanding state and temporary variables.

glass_a = "Water"
glass_b = "Milk"

print(f"Before: A={glass_a}, B={glass_b}")

# The Logic
temp_glass = glass_a   # Pour Water into Temp (Temp now has Water)
glass_a = glass_b      # Pour Milk into A (A now has Milk)
glass_b = temp_glass   # Pour Water (from Temp) into B (B now has Water)

print(f"After:  A={glass_a}, B={glass_b}")
```

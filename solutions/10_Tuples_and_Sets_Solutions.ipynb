```python
# Level 2: The Party List

guests = {"Tom", "Jerry", "Spike"}
guests.add("Tom") # Won't duplicate
guests.add("Tyke") # Will add
print(guests)
```

```python
# Level 3: Common Friends

friends_1 = {"Alice", "Bob", "Charlie"}
friends_2 = {"Bob", "David", "Edward"}
common = friends_1.intersection(friends_2)
print(f"Mutual friend: {common}")
```

```python
# Level 4: Grocery Cleaner

shop = ["Milk", "Bread", "Milk", "Eggs", "Bread"]
clean_shop = list(set(shop)) # One-liner trick!
print(clean_shop)
```

```python
# Level 5: The Skill Matcher

required = {"Python", "SQL"}
candidate = {"HTML", "CSS", "Python"}

matches = required.intersection(candidate)
print(f"Matched Skills: {matches}")

if len(matches) == len(required): # Or just == 2
    print("Hired!")
else:
    print("Missing Skills.")
```
